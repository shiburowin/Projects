```python
import random

player_1 = [random.randint(0, 25) for _ in range(5)]
print("player_1 values : ",player_1, "\n")
player_2 = [random.randint(0, 25) for _ in range(5)]
print("player_2 values : ",player_2, "\n")

while player_1 != [] and player_2 != []:
  for i in range(1):
    rolled_num = random.randint(0, 25)
    generated_num = []
    if rolled_num not in generated_num:
      generated_num.append(rolled_num)
    print(generated_num)

    for i in range(len(player_1)):
      if rolled_num in player_1:
        player_1.remove(rolled_num)
    print(player_1)

    for i in range(len(player_2)):
      if rolled_num in player_2:
        player_2.remove(rolled_num)
    print(player_2)

    if player_1 == [] and player_2 == []:
      print("Both  Wins \n")
    elif player_1 == []:
      print("Player_1 Wins \n ")
    elif player_2 == []:
      print("player_2 Wins \n")
    else:
      print("Still you have some numbers roll the dice again \n")
```
